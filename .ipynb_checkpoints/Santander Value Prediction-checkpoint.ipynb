## Importação das bibliotecas

```python
#Criação e manipulação dos dataframes
import pandas as pd

#Operações matemáticas
import numpy as np

#Separar o dataframe em treino e teste
from sklearn.model_selection import train_test_split, GridSearchCV

#Métricas de avaliação do modelo
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

#Regressores utilizados
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.neural_network import MLPRegressor 
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor 
from sklearn.svm import SVR
from sklearn.gaussian_process import GaussianProcessRegressor
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor

#Criação de Data Matrix para o XGBoost
import xgboost as xgb

#Normalização dos dados 
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

#Ignorar os avisos do sklearn
import warnings
warnings.filterwarnings('ignore')

#Verificar o tempo de run de cada modelo 
from time import time

#Visualização gráfica dos dados
import seaborn as sns
import matplotlib.pyplot as plt
```

## Verificar por valores nulos e colunas constantes (apenas zeros) no dataset

```python
def verificar_nulos(df):
    i = 0
    for coluna in df.columns:
        if df[coluna].isnull().sum() > 0:
            print(f'A coluna {coluna} possui {df[coluna].isnull().sum()} valores nulos.')
        
        if df[coluna].isnull().sum() == 0:
            i += 1
    
    if i == df.shape[1]:
        print('O dataframe não possui nenhum valor nulo')
        
def verificar_colunas_constantes(df):
    tamanho_original = df.shape
    df = df.loc[:, (df != df.iloc[0]).any()]
    
    tamanho_transformado = df.shape
    
    if tamanho_original != tamanho_transformado:
        print(f'Foram removidas {tamanho_original[1] - tamanho_transformado[1]} colunas constantes do dataframe.')
        
    else:
        print('O dataset não possui colunas constantes')
        
verificar_nulos(df)
verificar_colunas_constantes(df)
```

## Separar entre treino e teste

```python
y = df['target']
X = df.drop(['ID', 'target'], axis=1)

def separar_treino_teste(x,y, test_size):
    global X_train, X_test, y_train, y_test
    X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=test_size, random_state=42)
    
    print(f'Shape X_Train: {X_train.shape}')
    print(f'Shape y_Train: {y_train.shape}')
    print(f'Shape X_Test: {X_test.shape}')
    print(f'Shape y_Test: {y_test.shape}')
    
def separar_treino_teste_validacao(x,y, test_size):
    global X_train, X_test, X_validation, y_train, y_test, y_validation
    X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=test_size, random_state=42)
    X_test, X_validation, y_test, y_validation = train_test_split(X_test, y_test, test_size=0.5, random_state=42)
    
    print(f'Shape X_Train: {X_train.shape}')
    print(f'Shape y_Train: {y_train.shape}')
    print(f'Shape X_Test: {X_test.shape}')
    print(f'Shape y_Test: {y_test.shape}')
    print(f'Shape X_Validation: {X_validation.shape}')
    print(f'Shape y_Validation: {y_validation.shape}')

#Por já ter um dataset de teste separado, pode ser feito o split 90-10 entre treino e validação
def separar_treino_validacao(x,y, test_size):
    global X_train, X_validation, y_train, y_validation
    X_train, X_validation, y_train, y_validation = train_test_split(x, y, test_size=test_size, random_state=42)
    
    print(f'Shape X_Train: {X_train.shape}')
    print(f'Shape y_Train: {y_train.shape}')
    print(f'Shape X_Validation: {X_validation.shape}')
    print(f'Shape y_Validation: {y_validation.shape}')
    
separar_treino_teste_validacao(X, y, 0.20)
```

## Testando os diferentes tipos de escalonamento com alguns modelos

#### Como não foi observada uma mudança positiva nos melhores modelos utilizando os scalers, eles não são utilizados no modelo final

```python
scalers = [
    StandardScaler(),
    MinMaxScaler(),
    RobustScaler()
]

models = [
    LinearRegression(),
    Ridge(),
    MLPRegressor(),
    DecisionTreeRegressor(),
    KNeighborsRegressor()
]

for scaler in scalers:
    scaler = scaler.fit(X_train, X_test)
    X_train = scaler.transform(X_train)
    X_test = scaler.transform(X_test)
    
    print(f'Atual método utilizado: {scaler.__class__.__name__}')
    print(f'')
    
    for model in models:
        regressor = model.fit(X_train, y_train)
        y_pred = regressor.predict(X_test)
        print(f'O modelo {model.__class__.__name__} teve um MAE de {mean_absolute_error(y_test, y_pred)} e um r2 score de {r2_score(y_test, y_pred)}')
```

## Normalizando os dados com o Standard Scaler

```python
standard_scaler = StandardScaler().fit(X_train, X_test)
X_train = standard_scaler.transform(X_train)
X_test = standard_scaler.transform(X_test)
```

## Normalizando os dados com o MinMaxScaler

```python
minmaxscaler = MinMaxScaler().fit(X_train, X_test)
X_train = minmaxscaler.transform(X_train)
X_test = minmaxscaler.transform(X_test)
```

## Normalizando os dados com o Robust Scaler

```python
robustscaler = RobustScaler().fit(X_train, X_test)
X_train = robustscaler.transform(X_train)
X_test = robustscaler.transform(X_test)
```

## Testando todos os modelos de forma automatizada

```python
#Criando a lista dos modelos que serão testados
models = [
    LinearRegression(),
    Ridge(),
    Lasso(),
    RandomForestRegressor(),
    MLPRegressor(),
    DecisionTreeRegressor(),
    KNeighborsRegressor(),
    SVR(),
    GaussianProcessRegressor(),
    XGBRegressor(),
    CatBoostRegressor(),
    LGBMRegressor()
]

#Criando as listas vazias das métricas do modelo
MAE = []
RMSE = []
MSE = []
r2 = []
names = []
parameters = []
tempo = []

#Testando modelo por modelo
for model in models:
    #Startando o cronometro para ver o tempo de cada modelo 
    tempo_inicial = time()
    
    #Chamando o regressor e fazendo o predict
    regressor = model.fit(X_train, y_train)
    y_pred = regressor.predict(X_test)
    
    #Adicionando o valor da métrica de cada modelo para criar o dataframe depois
    MAE.append(mean_absolute_error(y_test, y_pred))
    RMSE.append(np.sqrt(mean_squared_error(y_test, y_pred)))
    MSE.append(mean_squared_error(y_test, y_pred))
    r2.append(r2_score(y_test, y_pred))
               
    #Para pegar o nome do modelo basta pegar o nome da classe
    names.append(model.__class__.__name__)
               
    parameters.append(model.get_params())
               
    #Parando o cronometro para ver o tempo e adicionando na lista 
    tempo_final = time() - tempo_inicial
    tempo.append(tempo_final)
    
#Chamando o cronometro para verificar o tempo de run do RandomForest com SelectKBest e GridSearchCV
tempo_inicial_randomforest = time()
    
#Utilizando o método SelectKBest para melhorar o modelo 
sel_kbest = SelectKBest(f_regression, k=45).fit(X_train, y_train)
X_train_sel = sel_kbest.transform(X_train)
X_test_sel = sel_kbest.transform(X_test)

#Lista de parâmetros para o RandomForest
randomforest_parameters = {'max_depth': [2,6,10,20,30,40],
             'n_estimators': [20,50,100,200,300,500],
             'max_features': ['sqrt', 'log2']}

#Desenvolvendo o modelo de Random Forest utilizando o SelectKBest e o GridSearch para encontrar os melhores parâmetros 
random_forest_regressor = GridSearchCV(RandomForestRegressor(), randomforest_parameters, n_jobs=-1)
random_forest_regressor.fit(X_train_sel, y_train)
y_pred_randomforest = random_forest_regressor.predict(X_test_sel)

#Parando de cronometrar o tempo do RandomForest e começando o do XGboost
tempo_final_randomforest = time() - tempo_inicial_randomforest
tempo_inicial_xgboost = time()

#Desenvolvendo o modelo de XGBoost utilizando o SelectKBest
xgb_regressor = XGBRegressor(n_estimators=1000, max_depth=10, eta=0.001, subsample=0.6,
                         colsample_bytree=0.6, random_state=42, eval_metric='rmse')
xgb_regressor.fit(X_train_sel, y_train)
y_pred_xgb = xgb_regressor.predict(X_test_sel)

#Parando o cronometro do xgboost 
tempo_final_xgboost = time() - tempo_inicial_xgboost

#Adicionando os valores do RandomForest e XGBoost para a lista do dataframe 
MAE.append(mean_absolute_error(y_test, y_pred_randomforest))
RMSE.append(np.sqrt(mean_squared_error(y_test, y_pred_randomforest)))
MSE.append(mean_squared_error(y_test, y_pred_randomforest))
r2.append(r2_score(y_test, y_pred_randomforest))
names.append('Random Forest GridSearchCV')
parameters.append(random_forest_regressor.best_params_)
tempo.append(tempo_final_randomforest)

#XGboost
MAE.append(mean_absolute_error(y_test, y_pred_xgb))
RMSE.append(np.sqrt(mean_squared_error(y_test, y_pred_xgb)))
MSE.append(mean_squared_error(y_test, y_pred_xgb))
r2.append(r2_score(y_test, y_pred_randomforest))
names.append('XGBoost SelectKBest')
parameters.append(xgb_regressor.get_params())
tempo.append(tempo_final_xgboost)


#Criando o dataframe com as métricas de todos os modelos 
resultado = pd.DataFrame({'Nome': names,
                         'Parametros': parameters,
                         'MAE': MAE,
                         'RMSE': RMSE,
                         'MSE': MSE,
                         'R2': r2,
                         'Run Time': tempo})
```

```python
resultado.sort_values(by='R2', ascending=False)
```

```python
regressor = CatBoostRegressor(iterations=1200, learning_rate=0.01, depth=10, eval_metric='RMSE',
                             random_seed=42, bagging_temperature=0.2, od_type='Iter', 
                             metric_period=50, od_wait=20)

regressor.fit(X_train, y_train, eval_set=(X_validation, y_validation), use_best_model=True, verbose=1)
y_pred = regressor.predict(X_test)
print(mean_squared_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```
